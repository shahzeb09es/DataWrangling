```python
# Dropping the unnecessary empty columns.
df.drop(['Unnamed: 2','Unnamed: 4','Unnamed: 6', 'Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9'],axis=1,inplace=True)
df.head(25)
```

```python
# Renaming the available columns with appropriate names
df.rename(columns={'Unnamed: 0':'Emp_ID','Unnamed: 1':'Emp_name','Unnamed: 3':'Location','Employee Previous Earnings':
                   'Description','Unnamed: 10':'Hours','Unnamed: 11':'Amount'},inplace=True)
df.head(25)
```

```python
# Employee Data Extraction Function 
def data_extraction_employees(emp1,emp2):
    #emp1=id[0] means first employee id
    #emp2=id[1] means second employee id

    fg=df['Emp_ID'][df['Emp_ID']==emp1].index #Finding the indices of first employee
    gf=df['Emp_ID'][df['Emp_ID']==emp2].index #Finding the indices of 2nd employee
    check=df['Emp_ID'].str.split("-",n=2,expand=False) #The first column contains the all conditional information
    
    # PayDates extraction loop based on conditions of Emp_ID column
    date=[] 

    for i in list(range(fg[0],gf[0]+1)):
        if not check.isna()[i]:
            if len(check[i]) == 3:
                a=check[i][2]
                if a in ['82','83', '84','85']:
                    date.append(check[i][0]+'-'+check[i][1]+'-'+check[i][2])
            
    date.append('Employee Total')     
    date.append('SA Payroll (Registered to Shahzeb CO. PTY. LTD.)')
            
    # Synchronising the Employee names and IDs based on total PayDates
    emp_id=[]
    emp_name=[]
    for i in list(range(len(date)-1)):
        emp=emp1
        emp_id.append(emp)
        name=df['Emp_name'][df['Emp_ID']==emp1].array[0]
        emp_name.append(name)

    dates=date.copy()
    dates.remove('SA Payroll (Registered to Shahzeb CO. PTY. LTD.)')

    # Generating the Dataframe based on each employee's own PayDate
    data_tuples = list(zip(emp_id,emp_name,dates))
    new_data=pd.DataFrame(data_tuples,columns=['Emp_ID','Emp_name','PayDate'])

    # Data manipulatioin and final Dataframe building of individual employee
    desc=[]
    hours=[]
    amount=[]
    for i in list(range(len(date)-1)):
        bt=df['Emp_ID'][fg[0]:gf[0]][df['Emp_ID']==date[i]].index
        gt=df['Emp_ID'][fg[0]:gf[0]][df['Emp_ID']==date[i+1]].index
        nm=gt[gt > bt[0]][0] 
        des=df['Description'][bt[0]+1:nm]
        hou=df['Hours'][bt[0]+1:nm]
        amo=df['Amount'][bt[0]+1:nm]
        desc.append(des)
        hours.append(hou)
        amount.append(amo)

    desc2=[]
    hour2=[]
    amount2=[]
    for i in list(range(len(desc))):
        fg=desc[i].array
        gh=hours[i].array
        hj=amount[i].array
        desc2.append(fg)
        hour2.append(gh)
        amount2.append(hj)

    data_tuples_vale = list(zip(desc2 , hour2, amount2))

    new_data_vale=pd.DataFrame(data_tuples_vale,columns=['Description','Hours','Amount'])
    new_data3=pd.concat([new_data,new_data_vale],axis=1)
    new_data3.stack(level=0,dropna=True).head(10)
    long_df=new_data3.copy()

    new_amount=[]
    new_desc=[]
    new_hours=[]
    for i in list(range(len(long_df['Amount']))):
        jn=long_df['Amount'][i].tolist()
        jh=long_df['Description'][i].tolist()
        jk=long_df['Hours'][i].tolist()
        new_amount.append(jn)
        new_desc.append(jh)
        new_hours.append(jk)

    long_df['Amount']=new_amount
    long_df['Description']=new_desc
    long_df['Hours']=new_hours
    return long_df
```

```python
# Extracting the all employees information
data=[]
for i in list(range(len(id)-1)):
    y=data_extraction_employees(emp1=id[i],emp2=id[i+1])
    data.append(y)
```

```python
# Concating the dataframes of all individual employee's paydates, hours and amounts
data=pd.concat(data)
data
```
